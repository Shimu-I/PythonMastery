## Exception Handling

zeroDivide.py

```python
def spam(divide_by):
  return 42 / divide_by

print(spam(2))
print(spam(12))
print(spam(0))
print(spam(1))
```

```python
def spam(divide_by):
  try:
   return 42 / divide_by
  except ZeroDivisionError:
   print('Error: Invalid argument.')
  

print(spam(2))
print(spam(12))
print(spam(0))
print(spam(1))
```

## A Short Program: Zigzag

zigzag.py

```python
import sys, time
indent = 0
indentIncreasing = True

try:
  while True:
    print(' ' * indent, end='')
    print('********')
    time.sleep(.1)

    if indentIncreasing:
      indent = indent + 1
      
      if indent == 20:
        indentIncreasing = False

    else:
      indent = indent - 1
      if indent == 0:
        indentIncreasing = True

except KeyboardInterrupt:
  sys.exit()
```

## Short Program: Spike

spike.py

```python
import time, sys

try:
  while True:
    for i in range(1, 9):
      print('-' * i * i)
      time.sleep(0.1)

    for i in range(7, 1, -1):
      print('-' * i * i)
      time.sleep(0.1)
except KeyboardInterrupt:
  sys.exit()
```
