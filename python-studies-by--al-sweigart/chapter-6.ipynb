## The List Data Type

```python
print([1, 2, 3])
print(['cat', 'bat', 'rat', 'elephant'])
print(['hello', 3.1415, True, None, 42])
spam = ['cat', 'bat', 'rat', 'elephant']
print(spam)
```

### Indexes


```python
spam = ['cat', 'bat', 'rat', 'elephant']
print(spam[0])
print(spam[1])
print(spam[2])
print(spam[3])

print(['cat', 'bat', 'rat', 'elephant'][3])
print('Hello, ' + spam[0])
print('The ' + spam[1] + ' ate the ' + spam[0] + '.')
```

```python
spam = ['cat', 'bat', 'rat', 'elephant']
print(spam[10000])
```

```python
spam = [['cat', 'bat'], [10, 20, 30, 40, 50]]
print(spam[0])
print(spam[0][1])
print([spam[1][4]])
```

### Negative indexes

```python
spam = ['cat', 'bat', 'rat', 'elephant' ]
print(spam[-1])
print(spam[-3])
print('The ' + spam[-1] + ' is afraid of the '+ spam[-3] + '.')
```

### Slices


```python
spam = ['cat', 'bat', 'rat', 'elephant']
print(spam[0:4])
print(spam[1:3])
print(spam[0:-1])
```

```python
spam = ['cat', 'bat', 'rat', 'elephant']
print(spam[:2])
print(spam[1:])
print(spam[:])
```

### The len() Function

```python
spam = ['cat', 'dog', 'moose']
print(len(spam))
```

### Value Update

```python
spam = ['cat', 'bat', 'rat', 'elephant']
spam[1] = 'aardvark'
print(spam)
spam[2] = spam[1]
print(spam)
spam[-1] = 12345
print(spam)
```

### Concatenation and Replication

```python

print([1, 2, 3] + ['A', 'B', 'C'])
print(['X', 'Y', "Z"] * 3)
spam = [1, 2, 3]
spam = spam + ['A', 'B', 'C']
print(spam)
```

### del Statements


```python
spam = ['cat', 'bat', 'rat', 'elephant']
del spam[2]
print(spam)
del spam[2]
print(spam)
```

## Working with Lists

allMyCats.py

```python
print('Enter the name of cat 1:')
cat_name_1 = input()
print('Enter the name of cat 2:')
cat_name_2 = input()
print('Enter the name of cat 3:')
cat_name_3 = input()
print('Enter the name of cat 4:')
cat_name_4 = input()
print('The cat names are:')
print(cat_name_1 + ' ' + cat_name_2 + ' ' + cat_name_3 + ' ' + cat_name_4)
```

allMyCats2.py

```python
cat_names = []
while True:
  print('Enter the name of cat ' + str(len(cat_names) + 1) + ' (Or enter nothing to stop.):')
  name = input()
  if name == '':
    break
  cat_names = cat_names + [name]
print('The cat names are:')
for name in cat_names:
  print(' ' + name)
```

### for Loops and Lists

```python
for i in range(4):
  print(i)
```

```python
for i in [1, 2, 3, 4]:
  print(i)
```

```python
supplies = ['pens', 'staplers', 'flamethrosers', 'binders']
for i in range(len(supplies)):
  print('Index ' + str(i) + ' in supplies is: ' + supplies[i])
```

### The in and not Operators

```python
print('howday' in ['hello', 'hi', 'howday', 'heyas'])

spam = ['hello', 'hi', 'howday', 'heyas']
print('cat' in spam)
print('howday' not in spam)
print('cat' not in spam)


```

myPets.py

```python
my_pets = ['Zophie', 'Pooka', 'Fat-tail']
print('Enter a pet name:')
name = input()
if name not in my_pets:
  print('I do not have a pet named ' + name)
else:
  print(name + ' is my pet.')
```

### The Multiple Assignment Trick

```python
cat = ['fat', 'gray', 'loud']
size = cat[0]
color = cat[1]
disposition = cat[2]

```

```python
## alternative way of writing
cat = ['fat', 'gray', 'loud']
size, color, disposition = cat
```

### List item Enumeration

```python
supplies = ['pens', 'staplers', 'flamethrowers', 'binders']
for index, item in enumerate(supplies):
  print('Index ' + str(index) + ' in supplies is: ' + item)
```

### Random Selection and Ordering

```python
import random

pets = ['Dog', 'Cat', 'Moose']

print(random.choice(pets))
print(random.choice(pets))
print(random.choice(pets))
```
