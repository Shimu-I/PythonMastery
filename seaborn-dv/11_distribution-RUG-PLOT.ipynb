# RUG PLOT

## Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import the supermart sales data excel file

```python
mart = pd.read_excel("/home/shimu/Documents/PythonMastery/seaborn-dv/supermarket_sales.xlsx")

mart = mart[['Gender', 'Payment', 'Unit price', 'Quantity', 'Total', 'gross income']]
mart.head(10)
```

## 1. Create a basic RUG plot on variable

```python
plt.figure(figsize=(15, 5))
sns.rugplot(data=mart, x='Unit price')
```

## 2. Create a RUG plot for two variable

```python
plt.figure(figsize=(15, 5))
sns.rugplot(data=mart, x='Unit price', y='gross income')
```

## 3. Group it by a catagorical variable using hue

```python
plt.figure(figsize=(15, 5))
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Gender')
```

## 4. Change the height of Rugs in RUG Plot

```python
plt.figure(figsize=(15, 5))
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Gender', height=.05)
```

## 5. Combine this with a KDE Plot

```python
plt.figure(figsize=(15, 5))
sns.kdeplot(data=mart, x='Unit price', y='gross income', hue='Gender', fill=True)
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Gender', height=.05)
```

## 6. Combine this with a SCATTER PLOT

```python
plt.figure(figsize=(15, 5))
sns.scatterplot(data=mart, x='Unit price', y='gross income', hue='Gender')
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Gender', height=.05)
```

## 7. Show Rugs outside of this Axis

```python
plt.figure(figsize=(15, 5))
sns.scatterplot(data=mart, x='Unit price', y='gross income', hue='Gender')
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Gender',  height=-0.04)
```
