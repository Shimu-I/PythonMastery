# DIS PLOT
# Distribution Plot

## 1. Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Import the supermart sales data excel file

```python
mart = pd.read_excel("/home/shimu/Documents/PythonMastery/seaborn-dv/supermarket_sales.xlsx")

mart = mart[['Gender', 'Payment', 'Unit price', 'Quantity', 'Total', 'gross income', 'Branch']]

mart.head(10)
```

## 1. Simple DIS plot

```python
# by default histragram
sns.displot(data=mart, x='Total')
```

## 2. With KDE

```python
sns.displot(data=mart, x='Total', kde=True)
```

## 3. With KDE and RUG

```python
sns.displot(data=mart, x='Total', kde=True, rug=True)
```

## 4. Change in RUG arguments

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={
    'height' : .05
})
```

## 5. Change in KDE arguments

```python
sns.displot(data=mart,
            x='Total',
            kde=True,
            rug=True,
            rug_kws={
                'height': .05
            },
            kde_kws={
                'bw_adjust':9
            }
            )
```

## 6. Show catagory

```python
sns.displot(data=mart, x='Total', kde=True,rug=True, hue='Payment')
```

```python
sns.displot(data=mart, x='Total', kde=True,rug=True, hue='Payment', multiple='stack')
```

## 7. Seperate Catagory

```python
sns.displot(data=mart, x='Total', kde=True,rug=True, hue='Payment', multiple='stack', col='Gender')
```

## 8. Add row

```python
sns.displot(data=mart, x='Total', kde=True,rug=True, hue='Payment', multiple='stack', col='Gender', row='Branch')
```

```python
sns.displot(data=mart, x='Total', kde=True,rug=True, hue='Payment', multiple='stack', col='Branch', row='Gender')
```

## 9. Make it BAR PLOT

```python
sns.displot(data=mart, y='Total', kde=True,rug=True, hue='Payment', multiple='stack', col='Branch', row='Gender')
```

## 10. Make it ECD plot

```python
sns.displot(data=mart, y='Total', kind='ecdf', hue='Payment', col='Branch', row='Gender')
```
