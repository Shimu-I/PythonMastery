# CLUSTER MAP PLOT

## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Datasest


```python
retail = pd.read_csv('/home/shimu/Documents/PythonMastery/seaborn-dv/sample_retail_dataset.csv')

retail.columns = retail.columns.str.lower()

retail.head()
```

## 1. Creating Pivot on the dataset

```python
retail_piv = retail.pivot_table(index='outlet_year',
                                columns='outlet_size',
                                values='sales')

retail_piv.head()
```

## 2. Create a basic cluster map

```python
# important for this

import scipy
print(scipy.__version__)
```

```python
sns.clustermap(retail_piv.fillna(0))
```

## 3. Remove column cluster

```python
sns.clustermap(retail_piv.fillna(0),
               col_cluster=False)
```

## 4. Remove row cluster

```python
sns.clustermap(retail_piv.fillna(0),
               row_cluster=False)
```

## 5. Add annotation

```python
sns.clustermap(retail_piv.fillna(0),
               col_cluster=False,
               annot=True)
```

```python
sns.clustermap(retail_piv.fillna(0),
               col_cluster=False,
               annot=True,
               fmt='.0f')
```

## 6. Change z_score 0 to 1

```python
sns.clustermap(retail_piv.fillna(0),
               col_cluster=False,
               annot=True,
               fmt='.1f',
               z_score=1)
```

## 7. Scale the plot

```python
sns.clustermap(retail_piv.fillna(0),
               col_cluster=False,
               annot=True,
               fmt='.1f',
               standard_scale=0)
```

## 8. Change line color

```python
sns.clustermap(retail_piv.fillna(0),
               col_cluster=False,
               annot=True,
               fmt='.1f',
               standard_scale=0,
               linewidth=.5,
               linecolor='red')
```
