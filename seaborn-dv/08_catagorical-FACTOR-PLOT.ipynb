# CAT PLOT

by default strip plot

kind='bar' --> allows to do different plot

## 1. Import all the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Import the supermart sales data excel file

```python
mart = pd.read_excel("/home/shimu/Documents/PythonMastery/seaborn-dv/supermarket_sales.xlsx")

mart.head()
```

## 3. Basic Cat plot (default is STRIP plot)

```python
sns.catplot(data=mart, x='Payment', y='Total')
```

## 4. BAR plot

```python
sns.catplot(data=mart, x='Payment', y='Total', kind='bar')
```

## 5. VIOLIN PLOT

```python
sns.catplot(data=mart, x='Payment', y='Total', hue='Gender', kind='violin')
```

## 6. Seperate based on gender 

```python
sns.catplot(data=mart, x='Payment', y='Total', hue='Gender', kind='violin', col='Gender')
```

## 7. Add new ROW

```python
sns.catplot(data=mart, x='Payment', y='Total', hue='Gender', kind='violin', col='Gender', row='Customer type')
```

```python
sns.catplot(data=mart, x='Payment', y='Total', hue='Gender', kind='violin', col='Gender', row='Customer type', color='gray')
```

```python
sns.catplot(data=mart, x='Payment', y='Total', hue='Gender', kind='violin', col='Gender', row='Customer type', palette='CMRmap_r')
```
