# ECDF
# Empirical Cumulative Distribution Functions

## Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##  Import the supermart sales data excel file

```python
mart = pd.read_excel("/home/shimu/Documents/PythonMastery/seaborn-dv/supermarket_sales.xlsx")

mart.head()
```

## 1. Create a simple ECDF Plot

```python
sns.ecdfplot(data=mart, x='gross income')
```

## 2. Plot based on catagory

```python
sns.ecdfplot(data=mart, x='gross income', hue='Gender')
```

## 3. Y axis variable change

```python
# by default the y = proportion
sns.ecdfplot(data=mart, x='gross income', hue='Gender', stat='count')
```

## 4. Plot in Y axis

```python
sns.ecdfplot(data=mart, y='gross income', hue='Gender', stat='count')
```
