# STRIP PLOT

A strip plot in Seaborn is a visualization used to show all individual data points for a variable across categories.

## 1. Import all the required libraried

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Import the supermart data excel file

```python
mart = pd.read_excel("/home/shimu/Documents/PythonMastery/seaborn-dv/supermarket_sales.xlsx")

mart.head()
```

## 3. Create a basic STRIP PLOT

```python
sns.stripplot(data=mart, y='Total')
```

## 4. Expand markers in STRIP PLOT using jitter

```python
sns.stripplot(data=mart, y='Total', jitter=.4)
```

## 5. Draw line around the points using LINEWIDTH

```python
sns.stripplot(data=mart, y='Total', linewidth=.3)
```

## 6. Include third catagoriecal variable with HUE

```python
sns.stripplot(data=mart, y='Total', x='Payment',hue='Gender', jitter=.2)
```

## 7. Separate each level of hue using DODGE

```python
sns.stripplot(data=mart, y='Total', x='Payment', hue='Gender', jitter=.2, linewidth=.2, dodge=True)
```

## 8. Draw the strips on top of VIOLIN PLOT

```python
sns.stripplot(data=mart, y='Total', x='Payment', jitter=.1)
sns.violinplot(data=mart, y='Total', x='Payment', color='gray')
```

## 9. Draw the strip on top of BOX PLOT

```python
sns.stripplot(data=mart, y='Total', x='Payment', jitter=.1)

sns.boxplot(data=mart, y='Total', x='Payment', color='gray')
```
