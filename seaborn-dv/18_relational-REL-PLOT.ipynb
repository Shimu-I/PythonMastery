# RELTIONAL PLOT
 by default scatter plot

##  Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import dataset

```python
fruits = pd.read_csv('/home/shimu/Documents/PythonMastery/seaborn-dv/fruits-dataset.csv')

fruits.head(10)
```

## 1. Create a basic REL Plot

```python
sns.relplot(data=fruits, x='outlet_year', y='sales')
```

## 1. Create a basic REL Plot with Kind attribute

```python
sns.relplot(data=fruits, x='outlet_year', y='sales', kind='line')
```

## 2. Create a basic REL Plot with kind and no Error line

```python
sns.relplot(data=fruits, x='outlet_year', y='sales', kind='line', errorbar=None)
```

## 3. Change catagory and size of marker

```python
sns.relplot(data=fruits, x='outlet_year', y='sales', s=100, hue='outlet_size')
```

## 4. Add column with feature

```python
sns.relplot(data=fruits, x='outlet_year', y='sales', s=100, hue='outlet_size', col='tier')
```

## 5. Add column with feature (with so many column)

```python
sns.relplot(data=fruits, x='outlet_year', y='sales', s=100, hue='outlet_size', col='item_type')
```

## 6. Add column with feature with column wrap

```python
sns.relplot(data=fruits, x='outlet_year', y='sales', s=100, hue='outlet_size', col='item_type', col_wrap=4)
```

## 7. Add row

```python
sns.relplot(data=fruits, x='outlet_year', y='sales', s=100, hue='outlet_size', col='outlet_size', row='tier')
```

## 8. Change marker color

```python
sns.relplot(data=fruits, x='outlet_year', y='sales', s=100, hue='outlet_size', palette=['green', 'red', 'yellow'])
```
