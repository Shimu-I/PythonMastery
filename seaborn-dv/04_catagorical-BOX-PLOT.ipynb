# BOX PLOT

A box plot (also called a box-and-whisker plot) is a graph that shows how data is spread out and where the middle values are.

---

📦 What it shows

It summarizes data using 5 key values:

Minimum → smallest value

Q1 (25%) → lower quarter

Median (50%) → middle value

Q3 (75%) → upper quarter

Maximum → largest value

---

📊 Structure (why it looks like a box)

Box → middle 50% of data (Q1 to Q3)

Line inside box → median

Whiskers → range (min to max, excluding outliers)

Dots (if any) → outliers (unusual values)

## 1. Import all the require LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Import MART dataframe

```python
mart = pd.read_excel('/home/shimu/Documents/PythonMastery/seaborn-dv/supermarket_sales.xlsx')

mart.head()
```

## 3. Create a basic BOX plot on one NUMERIC variable

```python
sns.boxplot(data=mart, y='Total' , width=.2)
```

## 4. Create a basic BOX plot on one numaric variable by CATAGORICAL variable

```python
sns.boxplot(data=mart, x='Payment', y='Total')
```

## 5. Create a basic BOX plot on one numaric variable by TWO CATAGORICAL variable using HUE attribute

```python
sns.boxplot(data=mart, x='Payment', y='Total', hue='Gender')
```

## 6. Add MEAN marker in the box plot using Showmeans attributes and change its style using meanprops

```python
sns.boxplot(data=mart, 
            x='Payment', 
            y='Total', 
            hue='Gender', 
            showmeans=True,
            meanprops={
                "marker": "o",
                "markerfacecolor": "white",
                "markersize": "8",
                "markeredgecolor": "black"
            })

# by default mean is false
```

## 7. Make HORIZONTAL box plot

```python
sns.boxplot(data=mart, 
            y='Payment', 
            x='Total', 
            hue='Gender', 
            showmeans=True,
            meanprops={
                "marker": "o",
                "markerfacecolor": "white",
                "markersize": "8",
                "markeredgecolor": "black"
            })

```

## 8. Change PALETTE, LINE WIDTH etc.

```python
sns.boxplot(data=mart, 
            x='Payment', 
            y='Total', 
            hue='Gender', 
            showmeans=True,
            meanprops={
                "marker": "o",
                "markerfacecolor": "white",
                "markersize": "8",
                "markeredgecolor": "black"
            },
            palette="CMRmap",
            linewidth=2)

```
