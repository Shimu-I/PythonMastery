# SCATTER PLOT

##  Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Dataset

```python
tips = sns.load_dataset('tips')
tips.head(10)
```

## 1. Create a basic SCATTER PLOT

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

## 2. Grouping basis on a catagorical variable using HUE

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
```

## 3. Grouping basis on a catagorical variable using STYLE ans Styling Markers a well

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='time')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='time', style='time', markers={
                    "Lunch": "^",
                    "Dinner": 'v'
                })
```

## 4. Grouping basis on a categorical variable using HUE and STYLE both together

```python
sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='day', style='time', markers={
                    "Lunch": "^",
                    "Dinner": 'v'
                })
```

## 5. Grouoping basis on numeric variable using HUE and use PALETTE

```python
sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='time', style='time', markers={
                    "Lunch": "^",
                    "Dinner": 'v'
                },
                palette='Purples_r')
```

## 6. Grouping basis on numeric variable using HUE and SIZE

```python
sns.scatterplot(data=tips, 
                x='total_bill', 
                y='tip',
                hue='time', 
                size='size',
                style='time', markers={
                    "Lunch": "^",
                    "Dinner": 'v'
                })
```

```python
sns.scatterplot(data=tips, 
                x='total_bill', 
                y='tip',
                hue='time', 
                size='size',
              
                sizes=(20, 200),
                style='time', markers={
                    "Lunch": "^",
                    "Dinner": 'v'
                })
```

## 7. Change the Marker size with S argument

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='time', s=100, edgecolor='black')
```
