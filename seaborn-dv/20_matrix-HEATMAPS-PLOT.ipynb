## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Datasest


```python
retail = pd.read_csv('/home/shimu/Documents/PythonMastery/seaborn-dv/sample_retail_dataset.csv')

retail.columns = retail.columns.str.lower()

retail.head()
```

```python
retail.shape
```

```python
retail_piv = retail.pivot_table(index='outlet_year',
                                columns='outlet_size',
                                values='sales')

retail_piv
```

## 1. Basic Heat map

```python
sns.heatmap(data=retail_piv)
```

## 2. Add figsize

```python
sns.set_style('white')
plt.figure(figsize=(5, 5))

sns.heatmap(data=retail_piv)
```

## 3. Style in square shape

```python
sns.set_style('white')
plt.figure(figsize=(5, 5))

sns.heatmap(data=retail_piv,
            square=True)
```

## 4. Populate Numbers

```python
sns.set_style('white')
plt.figure(figsize=(5, 5))

sns.heatmap(data=retail_piv,
            annot=True)
```

```python
sns.set_style('white')
plt.figure(figsize=(5, 5))

sns.heatmap(data=retail_piv,
            annot=True,
            fmt='.0f')
```

## 5. Make the Annotate number bigger

```python
sns.set_style('white')
plt.figure(figsize=(5, 5))

sns.heatmap(data=retail_piv,
            annot=True,
            fmt='.0f',
            annot_kws=dict(
                size=15,
                weight='bold'
                
            ))
```

## 6. Change line width and color

```python
sns.set_style('white')
plt.figure(figsize=(5, 5))

sns.heatmap(data=retail_piv,
            annot=True,
            fmt='.0f',
            linewidths=.5,
            linecolor='red',
            cmap='OrRd_r',
            annot_kws=dict(
                size=15,
                weight='bold'
               

            ))
```

## 7. See data with co-relation

```python
sns.heatmap(retail_piv.corr())
```

```python
sns.heatmap(retail_piv.corr(),
            vmin=-1,
            vmax=1,
            center=0)
```

## 8. CMAP change

```python
sns.heatmap(retail_piv.corr(),
            vmin=-1,
            vmax=1,
            center=0,
            cmap='OrRd_r')
```

```python
sns.heatmap(retail_piv.corr(),
            vmin=-1,
            vmax=1,
            center=0,
            cmap='OrRd_r',
            annot=True)
```

```python
sns.heatmap(retail_piv.corr(),
            vmin=-1,
            vmax=1,
            center=0,
            cmap='OrRd_r',
            annot=True,
            fmt='.1f')
```

```python
sns.heatmap(retail_piv.corr(),
            vmin=-1,
            vmax=1,
            center=0,
            cmap='OrRd_r',
            annot=True,
            fmt='.1f',
            linecolor='black',
            linewidths=.5,
            annot_kws=dict(
                size=15, weight='bold'
            ))
```
