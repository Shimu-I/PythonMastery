# JOINT PLOT


##  Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##  Import the supermart sales data excel file

```python
iris = sns.load_dataset('iris')

iris.head()
```

## 1. Creating a basic JOINT plot


```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length')
```

## 2. Changing the Plot kind to "scatter" | "kde" | "hist" | "hex" | "reg" | "resid"

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length', kind='scatter')
```

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length', kind='kde')
```

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length', kind='hist')
```

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length', kind='hex')
```

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length', kind='reg')
```

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length', kind='resid')
```

## 3. Groupting based on a catagorical variable

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length', hue='species')
```

## 4. Formatting All and Individual Plots

### ON ENTIRE PLOT

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length',)
```

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length', color='red')
```

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length',hue='species')
```

```python
sns.jointplot(data=iris, 
              x='sepal_length', 
              y='petal_length',
              hue='species',
              palette='BuPu'
              )
```

### ON INDIVIDUAL PLOT

```python
sns.jointplot(data=iris, 
              x='sepal_length', 
              y='petal_length',
              joint_kws=dict(
                  marker = '+',
                  color='red'
              )
              )
```

```python


sns.jointplot(data=iris, 
              x='sepal_length', 
              y='petal_length',
              joint_kws=dict(
                  marker = '+',
                  color='red'           
                  
              ),
              marginal_kws=dict(
                      color='green'
                  )
              )

```

```python


sns.jointplot(data=iris, 
              x='sepal_length', 
              y='petal_length',
              joint_kws=dict(
                  marker = '+',
                  color='red'           
                  
              ),
              marginal_kws=dict(
                      color='green',
                      kde=True
                  )
              )

```

## 5. Ajust height, ratio, space and showhide the marginal_ticks

```python


sns.jointplot(data=iris, 
              x='sepal_length', 
              y='petal_length',
              height=5
              )

```

```python


sns.jointplot(data=iris, 
              x='sepal_length', 
              y='petal_length',
              height=5,
              ratio=2
              )

```

```python
sns.jointplot(data=iris, 
              x='sepal_length', 
              y='petal_length',
              height=5,
              ratio=2,
              space=1
              )

```

```python


sns.jointplot(data=iris, 
              x='sepal_length', 
              y='petal_length',
              height=5,
              ratio=2,
              marginal_ticks=True
              )

```

## 6. Plot KDE and RUG on top of JOINT PLOT


```python
l = sns.jointplot(data=iris, 
              x='sepal_length', 
              y='petal_length'
              )

```

```python
l = sns.jointplot(data=iris, 
              x='sepal_length', 
              y='petal_length',
              color='red'
              )

l.plot_joint(sns.kdeplot)
```

```python
l = sns.jointplot(data=iris, 
              x='sepal_length', 
              y='petal_length',
              color='red'
              )

l.plot_joint(sns.rugplot, height=.1, color='green')
```
