# FACET GRID PLOT

## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Datasest


```python
retail = pd.read_csv('/home/shimu/Documents/PythonMastery/seaborn-dv/sample_retail_dataset.csv')

retail.columns = retail.columns.str.lower()

retail.head()
```

## 1. Cheack out the unique item

```python
retail.outlet_size.unique()
```

```python
retail.outlet_location_type.unique()
```

## 2. Create a basic plot

```python
sns.FacetGrid(retail, col='outlet_size')
```

## 3. Change to Histogram

```python
new = sns.FacetGrid(retail, col='outlet_size')

new.map_dataframe(sns.histplot, x='sales')
```

## 4. Change to KDE Plot

```python
new = sns.FacetGrid(retail, col='outlet_size')

new.map_dataframe(sns.kdeplot, x='sales')
```

## 5. Change to Line plot

```python
new = sns.FacetGrid(retail, col='outlet_size')

new.map_dataframe(sns.lineplot, x='outlet_year', y='sales')
```

## 6. Scatter plot

```python
new = sns.FacetGrid(retail, col='outlet_size')

new.map_dataframe(sns.scatterplot, x='outlet_year', y='sales')
```

## 7. Add styling

```python
new = sns.FacetGrid(retail, col='outlet_size')

new.map_dataframe(sns.scatterplot, x='outlet_year', y='sales',
                  hue='outlet_location_type')
```

```python
new = sns.FacetGrid(retail, col='outlet_size')

new.map_dataframe(sns.scatterplot, x='outlet_year', y='sales',
                  hue='outlet_location_type',
                  marker='^',
                  alpha=.5,
                  color='red',
                  s=100)
```

## 8. ADD row and column

```python

mew = sns.FacetGrid(retail, col='outlet_size',
                    row='outlet_location_type')
```

## 9. show Line plot in 5 plots

```python
new = sns.FacetGrid(retail, col='outlet_size',
                    row='outlet_location_type')

new.map_dataframe(sns.lineplot, x='outlet_year', y='sales')
```

## 10. Add seperate y axis for each

```python
new = sns.FacetGrid(retail, col='outlet_size',
                    row='outlet_location_type',
                    sharey=False)

new.map_dataframe(sns.lineplot, x='outlet_year', y='sales')
```

## 11. ALL plot y axis has same limit

```python
new = sns.FacetGrid(retail, col='outlet_size',
                    row='outlet_location_type',
                    sharey=False,
                    ylim=(0, 5000))

new.map_dataframe(sns.lineplot, x='outlet_year', y='sales')
```

## 12. ADD axis label for X and Y

```python
new = sns.FacetGrid(retail, col='outlet_size',
                    row='outlet_location_type',
                    sharey=False,
                    ylim=(0, 5000))

new.map_dataframe(sns.lineplot, x='outlet_year', y='sales')
new.set_axis_labels('sales', 'count')
```

## 13. Fix the text cluster 
## Change the column template

```python
new = sns.FacetGrid(retail, col='outlet_size',
                    row='outlet_location_type',
                    sharey=False,
                    ylim=(0, 5000))

new.map_dataframe(sns.lineplot, x='outlet_year', y='sales')
new.set_axis_labels('sales', 'count')
new.set_titles(col_template='{col_name}')
```

## 14. Change row template

```python
new = sns.FacetGrid(retail, col='outlet_size',
                    row='outlet_location_type',
                    sharey=False,
                    ylim=(0, 5000))

new.map_dataframe(sns.lineplot, x='outlet_year', y='sales')
new.set_axis_labels('sales', 'count')

new.set_titles(col_template='{col_name}', row_template='{row_name}')

```

```python
new = sns.FacetGrid(retail, col='outlet_size',
                    row='outlet_location_type',
                    sharey=False,
                    ylim=(0, 5000))

new.map_dataframe(sns.lineplot, x='outlet_year', y='sales')
new.set_axis_labels('sales', 'count')

new.set_titles(col_template='{col_name} Size', row_template='{row_name} Type')

```
