# SWARM PLOT

## 1. Import all the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Import the supermart sales data excel file

```python
mart = pd.read_excel("/home/shimu/Documents/PythonMastery/seaborn-dv/supermarket_sales.xlsx")

mart.head()
```

## 3. Create a basic SWARM PLOT on Total column

```python
sns.swarmplot(data=mart, y='Total')
```

## 4. Create SWARM PLOT group by CATEGORIES

```python
sns.swarmplot(data=mart, y='Total', x='Payment', hue='Gender')
```

## 5. Showing Hues Separately on the catagorical axis

```python
sns.swarmplot(data=mart, y='Total', x='Payment', hue='Gender', dodge=True)
```

## 6. Stying s SWARM - Change the matrix, size, color, edge color...

```python
plt.figure(figsize=(15, 5))
sns.swarmplot(data=mart, y='Total', x='Payment', marker="o", size=10, color='green', edgecolor='black')
```

## 7. Overlay a SWARM PLOT a BOX PLOT

```python
sns.swarmplot(data=mart, y='Total', x='Payment', color='black')
sns.boxplot(data=mart, y='Total', x='Payment')
```

## 8. Overlay a SWARM PLOT on a VIOLIN PLOT

```python
sns.swarmplot(data=mart, y='Total', x='Payment', color='white')
sns.violinplot(data=mart, y='Total', x='Payment', inner=None, width=1)
```
