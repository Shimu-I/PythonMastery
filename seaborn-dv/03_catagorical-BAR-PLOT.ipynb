# BAR PLOT

A bar plot shows the relationship between a category and a numerical value.

X-axis → categories (e.g., product types)
Y-axis → numbers (e.g., sales, average, sum)

👉 It is used when you already have a value to measure.

Statistical aggregation means the process of combining multiple individual data values into a summarized form using statistical functions.

In simpler terms, it’s when you take a bunch of raw data and compress it into meaningful summary statistics.

Common examples of aggregation:
Sum → total sales = 100 + 200 + 300 = 600
Average (mean) → average score = (70 + 80 + 90) / 3 = 80
Count → number of students = 50
Min / Max → lowest or highest value in a dataset
Median → middle value when data is sorted

## 4. Create a basic BARPLOT

```python
plt.figure(figsize=(15, 5))

sns.barplot(data=mart, x='Product line', y='Total')
```

The black error bar is there to show you the spread and uncertainty of that data.


The Top of the Blue Bar: The exact average estimate (e.g., 335).

The Very Top of the Black Line: The highest realistic estimate for the average (around 380).

The Very Bottom of the Black Line: The lowest realistic estimate for the average (around 300).

## 12. Using SATURATION parameter

Saturation of color (simple meaning)

Saturation means the intensity or purity of a color.

🔵 High saturation
Color looks strong, deep, and vivid
Example: bright red, rich blue

👉 No mixing with gray
⚪ Low saturation
Color looks faded, dull, or washed out
More gray mixed in

```python
plt.figure(figsize=(15, 5))

sns.barplot(data=mart, x='Product line', y='Total', order=['Electronic_accessories',    'Fashion_accessories',     'Food_and_beverages',
      'Health_and_beauty',     'Home_and_lifestyle',      'Sports_and_travel'],
      hue='Gender',
      palette='cividis',
      saturation=.5
      )
```

## CHANGE DEFAULT AGGREGATION METHOD USING ESTIMATOR PARAMETER

-- by default is mean


```python
plt.figure(figsize=(15, 5))

sns.barplot(data=mart, x='Product line', y='Total', order=['Electronic_accessories',    'Fashion_accessories',     'Food_and_beverages',
      'Health_and_beauty',     'Home_and_lifestyle',      'Sports_and_travel'],
      hue='Gender',
      palette='cividis',
      saturation=.5,
      estimator=np.median
      )
```
