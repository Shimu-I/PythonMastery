# LINE PLOT

##  Import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Import Dataset

```python
tips = sns.load_dataset('tips')

tips.head(15)
```

## 1.Create a basic line plot and try different estimators


```python
sns.lineplot(data=tips, x='total_bill', y='tip')
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', errorbar=None)
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', errorbar=None, estimator=sum)
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', errorbar=None, estimator=None)
```

```python
#------Estimator = None

height = [5, 5, 7, 5, 8]
age = [22, 42, 30, 35, 40]

sns.lineplot(x=age, y=height, estimator=None)
```

```python
#----- Estimator = sum

height = [5, 6, 7, 6.5, 4]
age = [22, 25, 30, 25, 25]

sns.lineplot(x=age, y=height, estimator=sum)
```

## 2. Test with different Confidence intervals and with different number of Bootstraping

```python
# standard deviation of mean
sns.lineplot(data=tips, x='total_bill', y='tip', errorbar='sd')
```

```python
# standard deviation of mean
sns.lineplot(data=tips, x='total_bill', y='tip', errorbar=('ci', 20))
```

```python
# by default bookstrape is 1000
sns.lineplot(data=tips, x='total_bill', y='tip', errorbar=('ci', 20), n_boot=400)
```

## 3. Grouping using HUE and use Different Palettes

```python

sns.lineplot(data=tips, x='total_bill', y='tip', hue='day')
```

## 4. Grouping using Style, use different Marker and Dashes

```python
sns.lineplot(data=tips, x='day', y='total_bill', hue='day',
             style='day',
             marker=True,
             dashes=False)
```

## 5. Grouping using Size

```python
sns.lineplot(data=tips, x='total_bill', y='tip',
             size='total_bill',
             sizes=(.5, 10))
```

## 6.Use different semantic styling parameters on same or different variables



```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='day',markers=True)
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day',  style='sex',markers=True)
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day',  style='sex',markers=True, size='day')
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day',  style='sex',markers=True, size='size')
```
