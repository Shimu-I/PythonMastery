# KDE PLOT
# Kernal Density Estimation Plot

## 1. Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Import the supermart sales data excel file

```python
mart = pd.read_excel("/home/shimu/Documents/PythonMastery/seaborn-dv/supermarket_sales.xlsx")

mart.head(10)
```

## 1. Create a basic KDE plot for Total column

```python
sns.kdeplot(data=mart, x='Total')
```

## 2. Create KDE for all the numeric variables in dataframe

```python
sns.kdeplot(data=mart)
```

## 3. Adjust the smoothing using bw_adjust

```python
sns.kdeplot(data=mart, x='Total')
```

```python
sns.kdeplot(data=mart, x='Total', bw_adjust=.5)
```

## 4. Group the KDE on a catagory variable

```python
sns.kdeplot(data=mart, x='Total', hue='Payment')
```

## 5. Stack KDE on a catagory using MULTIPLE argument

```python
sns.kdeplot(data=mart, x='Total', hue='Payment', multiple='stack')
```

## 6. Use log scaling to map the variable in KDE

```python
sns.kdeplot(data=mart, x='Total', log_scale=True)
```

## 7. Change styling of hued KDE using linewidth, palette, alpha etc...


```python
sns.kdeplot(data=mart, x='Total', hue='Payment', multiple='stack', linewidth=5)
```

```python
sns.kdeplot(data=mart, x='Total', hue='Payment', multiple='stack', linewidth=5, palette='gray', alpha=.2)
```

## 8. Create a bivariate KDE

```python
sns.kdeplot(data=mart, x='Unit price', y='gross income')
```

## 9. Group the bivarinate KDE on a categorical variable and show the contours

```python
sns.kdeplot(data=mart, x='Unit price', y='gross income', hue='Gender')
```

```python
sns.kdeplot(data=mart, x='Unit price', y='gross income', hue='Gender', fill=True)
```

```python
sns.kdeplot(data=mart, x='Unit price', y='gross income', hue='Gender', fill=True, levels=5)
```

```python
sns.kdeplot(data=mart, x='Unit price', y='gross income', hue='Gender', fill=True, thresh=.2)
```
