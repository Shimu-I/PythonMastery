# HISTOGRAM PLOT

## Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import the supermart sales data excel file

```python
mart = pd.read_excel("/home/shimu/Documents/PythonMastery/seaborn-dv/supermarket_sales.xlsx")

mart.head()
```

## 1. Create a basic HISTROGRAM on X and Y axiz

```python
sns.histplot(data=mart, x='Total')
```

```python
sns.histplot(data=mart, y='Total')
```

## 2. Change the Bins Width using binwidth argument

```python
sns.histplot(data=mart, x='Total', binwidth=150)
```

## 3. Change Number of Bins and interval

```python
sns.histplot(data=mart, 
             x='Total',
             bins=30)
```

```python
plt.figure(figsize=(15, 5))
sns.histplot(data=mart, 
             x='Total'
             )
plt.xticks(np.arange(0, 1100, 50))
```

## 4. Combine with KDE

```python
sns.histplot(data=mart, x='Total', kde=True)
```

## 5. Use a categorical variable in HUE and STACK it using MULTIPLE argument

```python
sns.histplot(data=mart, x='Total', hue='Payment', multiple='stack')
```

## 6. Make it a step/poly plot using ELEMENT argument and change the FILL

```python
sns.histplot(data=mart, x='Total', element='step')
```

```python
sns.histplot(data=mart, x='Total', element='step', fill=False)
```

```python
sns.histplot(data=mart, x='Total', element='poly', fill=False)
```

```python
sns.histplot(data=mart, x='Total',hue='Gender', element='poly')
```

## 7. Use categorical variable and shrink it

```python
sns.histplot(data=mart, x='Payment')
```

```python
sns.histplot(data=mart, x='Payment', shrink=.7)
```

```python
sns.histplot(data=mart, x='Total', stat='probability')
```

## 8. Create a Biariate Histogram

```python
sns.histplot(data=mart, x='Total', y='gross income')
```
