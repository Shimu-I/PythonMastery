# VIOLIN PLOT

A violin plot in Seaborn is a statistical visualization used to show the distribution of numeric data across categories.

 = 


## 1. Import all the required libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Import the supermart data excel file

```python
mart = pd.read_excel('/home/shimu/Documents/PythonMastery/seaborn-dv/supermarket_sales.xlsx')

mart.head()
```

## 3. Create a basic VIOLIN PLOT

```python
sns.violinplot(data=mart,
               y='Total')
```

## 4. Create a VIOLIN PLOT on two catagorical and one numaric variable and use split

```python
sns.violinplot(data=mart,
               x='Payment', 
               y='Total',
               hue='Gender',
               split=True)
```

## 5. Change the box in the VIOLIN PLOT to horizontal lines

```python
sns.violinplot(data=mart,
               x='Payment', 
               y='Total',
               hue='Gender',
               split=True,
               inner='quartile')
```

## 6. Draw line for each observation in a VIOLIN PLOT

```python
sns.violinplot(data=mart,
               x='Payment', 
               y='Total',
               hue='Gender',
               split=True,
               inner='stick')
```

## 7. Change the amount of smoothing using bw attribute

```python
sns.violinplot(data=mart,
               x='Payment', 
               y='Total',
               hue='Gender',
               split=True,
               inner='quartile',
               bw=.2
               )
```

## 8. Cut out the extreme values


```python
sns.violinplot(data=mart,
               x='Payment', 
               y='Total',
               hue='Gender',
               split=True,
               inner='quartile',
               bw=.2,
               cut=0
               )
```
