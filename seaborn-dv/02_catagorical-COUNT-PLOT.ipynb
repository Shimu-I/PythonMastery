# COUNT PLOT

A count plot shows the number of occurrences of each category.

X-axis → categories
Y-axis → how many times each category appears

👉 It is used when you just want to count data, not calculate values.

## 1. Import all the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Import the data

```python
mart = pd.read_excel("/home/shimu/Documents/PythonMastery/seaborn-dv/supermarket_sales.xlsx")

mart
```

## 3. Create a basic COUNTPLOT to get number of transaction for each of the product line and change the figure size

```python
plt.figure(figsize=(15,5))
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Product line', data=mart)
```

## 4. Make it horizontal bar plot

```python
sns.countplot(y='Product line', data=mart)
```

## 5. Add hue to get the count on two categorical i.e. Product line and Gender

```python
plt.figure(figsize=(15, 5))
sns.countplot(data=mart, x='Product line', hue='Gender')
```

## 6. Using different color palette

```python


print(sns.palettes.SEABORN_PALETTES.keys())
```

```python
plt.figure(figsize=(15, 5))
sns.countplot(data=mart, x='Product line', hue='Gender', palette='deep')
# asdfa
```

## 7.Change Style using facecolor, linewidth and edge color

```python
plt.figure(figsize=(15, 5))

sns.countplot(data=mart, 
              x='Product line',
              facecolor= (1,0,1,0),
              linewidth=8,
              edgecolor=sns.color_palette('dark', 3))
```
