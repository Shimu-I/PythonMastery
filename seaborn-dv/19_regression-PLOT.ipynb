# REGRESSION PLOT



## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Dataset

```python
tips = sns.load_dataset('tips')

tips.head()
```

## 1. Create a basic reg plot

```python
sns.regplot(data=tips, x='total_bill', y='tip')
```

## 2. Change the styling of reg plot, only line, only scatter, show/hide the errorbar, change the ci value, change n_boot

```python
sns.regplot(data=tips, x='total_bill', y='tip', color='green', marker='^')
```

```python
sns.regplot(data=tips, x='total_bill', y='tip', line_kws=dict(
    color='red',
    linestyle='--'
),
scatter_kws=dict(
    s=100, color='green',
    alpha=.5
))
```

```python
sns.regplot(data=tips, x='total_bill', y='tip',
    line_kws=dict(
    color='red',
    linestyle='--'),
scatter_kws=dict(
    s=100, color='green',
    alpha=.5
),
)
```
